```python
import os
import json
import pickle

from model.state import InteractionData
from log.string import print_interaction_data

all_interaction_data = InteractionData()
for file in os.listdir('./data/metalarcwelding/expert_sessions'):
    with open(f'./data/metalarcwelding/expert_sessions/{file}', 'r') as f:
        interaction_data = json.load(f)
        all_interaction_data.sessions.append(InteractionData.parse_obj(interaction_data).sessions[0])

print_interaction_data(all_interaction_data)

with open("./data/metalarcwelding/interaction/interaction_data.pkl", "wb") as f:
    pickle.dump(all_interaction_data, f)
```

```python
from log.string import print_abstract_interaction_sequences
from guidance.filter import filter_segments
import pickle

with open("./data/metalarcwelding/interaction/segmented_data.pkl", "rb") as f:
    segmented_data = pickle.load(f)

filtered_segments = filter_segments(segmented_data)
print_abstract_interaction_sequences(filtered_segments)

with open("./data/metalarcwelding/interaction/filtered_segments.pkl", "wb") as f:
    pickle.dump(filtered_segments, f)
```

```python
from log.string import print_abstract_interaction_sequence_clusters
from guidance.clustering import cluster_interaction_data
import pickle

with open("./data/metalarcwelding/interaction/filtered_segments.pkl", "rb") as f:
    filtered_segments = pickle.load(f)

numer_of_sessions = 4
eps = 0.2
clusters = cluster_interaction_data(filtered_segments, numer_of_sessions, eps)
outliers = clusters.pop('-1', None)
print_abstract_interaction_sequence_clusters(clusters)

if outliers is not None:
    print("---------------------------------")
    print(f"Removed {len(outliers.sequences)} outlier(s)")

with open("./data/metalarcwelding/interaction/clusters.pkl", "wb") as f:
    pickle.dump(clusters, f)
```

```python
from guidance.recurring_pattern import Pattern
from guidance.clustering import Cluster
from typing import Dict, List
from log.string import print_abstract_interaction_cluster_patterns
import pickle

with open("./data/metalarcwelding/interaction/clusters.pkl", "rb") as f:
    clusters: Dict[str, Cluster] = pickle.load(f)

print_abstract_interaction_cluster_patterns(clusters)
patterns: List[Pattern] = []
for key, cluster in clusters.items():
    pattern = Pattern()
    pattern.score = cluster.score
    pattern.pattern = cluster.most_rep
    patterns.append(pattern)

patterns.sort(key=lambda x: x.score)

with open("./data/metalarcwelding/interaction/recurring_patterns.pkl", "wb") as f:
    pickle.dump(patterns, f)
```
